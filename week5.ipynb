```python
age = int(input('Age: '))

if age >= 60:
    print("You are a senior citizen.")
else:
    if age >= 18:
        print("You are an adult.")
    else:
        print("You are a teenager.")
```

```python
tahun = int(input('Tahun: ' ))


if (tahun % 400 == 0):
  print(f"{tahun} adalah tahun kabisat")
else:
  if (tahun % 100 == 0):
      print(f"{tahun} bukan tahun kabisat.")
  else:
      if (tahun % 4 == 0):
        print (f"{tahun} adalah tahun kabisat")
      else:
        print(f"{tahun} bukan tahun kabisat.")
```

```python
angka_pertama = int(input('Masukkan Angka Pertama: '))
angka_kedua = int(input('Masukkan Angka Kedua: '))
angka_ketiga = int(input('Masukkan Angka Ketiga: '))

print(f"Angka pertama: {angka_pertama}")
print(f"Angka kedua: {angka_kedua}")
print(f"Angka ketiga: {angka_ketiga}")

if angka_pertama >= angka_kedua and angka_pertama >= angka_ketiga:
  print(f"Angka Terbesar adalah: {angka_pertama}")
else:
  if angka_kedua >= angka_pertama and angka_kedua >= angka_ketiga:
    print(f"Angka Terbesar adalah: {angka_kedua}")
  else:
    print(f"Angka Terbesar adalah: {angka_ketiga}")
```

```python
harga = int(input('Total Belanjaan: '))
member = int(harga * 2 / 100)

if harga >= 200000:
  diskon = harga * 4 / 100
  print(f"Member harus membayar sebesar {harga - diskon - member}")
  print(f"Bukan member harus membayar sebesar {harga - diskon}")
else:
    if harga >= 100000:
      diskon = harga * 3 / 100
      print(f"Member harus membayar sebesar {harga - diskon - member}")
      print(f"Bukan member harus membayar sebesar {harga - diskon}")
    else:
      print(f"Member harus membayar sebesar {harga - member}")
      print(f"Bukan member harus membayar sebesar {harga}")
```

```python
username = input("Username: ")
email = input("Email: ")
password = input("Password: ")

if (username == ""):
  print("Username tidak boleh kosong.")
else:
  if "@" not in email:
    print("Email harus memuat @")
  elif len(password) < 8:
    print("Password harus lebih dari 8 karakter")
  else:
    print("User berhasil dibuat")
```
